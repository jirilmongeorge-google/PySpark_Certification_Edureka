### The Explode Function

```python
%%read_sql df -c QS_CONNECTION
create table test as 
select
ot_Vid,
orderid,
products.productLabels.product_ProductId as productid,
products.attributedLabels.order_Quantity as order_quantity,
products.calculatedProductLabels.units as units,
products.calculatedProductLabels.revenue as revenue
FROM   (SELECT _ottoaeppoc.ot_Vid as ot_vid, 
        Explode(_ottoaeppoc.productContext.products)   AS products,
        _ottoaeppoc.baseLabels.order_OrderId as orderid
from otto_develop)
limit 50;
```

```python
%%read_sql df -c QS_CONNECTION
create view table_view_for_demo as
select
ot_Vid,
orderid,
products.productLabels.product_ProductId as productid,
products.attributedLabels.order_Quantity as order_quantity,
products.calculatedProductLabels.units as units,
products.calculatedProductLabels.revenue as revenue
FROM   (SELECT _ottoaeppoc.ot_Vid as ot_vid, 
        Explode(_ottoaeppoc.productContext.products)   AS products,
        _ottoaeppoc.baseLabels.order_OrderId as orderid
from otto_develop)
limit 50;
```

```python
%%read_sql df -c QS_CONNECTION
select
ot_Vid,
orderid,
products.productLabels.product_ProductId as productid,
products.attributedLabels.order_Quantity as order_quantity,
products.calculatedProductLabels.units as units,
products.calculatedProductLabels.revenue as revenue,
products.productLabels.product_VariationId as variationid,
basketid,
onex_Experiments
FROM   (SELECT _ottoaeppoc.ot_Vid as ot_vid, 
        Explode(_ottoaeppoc.productContext.products)   AS products,
        _ottoaeppoc.baseLabels.order_OrderId as orderid,
        _ottoaeppoc.baseLabels.order_BasketId as basketid,
        _ottoaeppoc.listLabels.onex_Experiments as onex_Experiments
from DataSet_OTTO_IterationFinal2)
limit 1000;
```

```python
%%read_sql data -c QS_CONNECTION
select
basketid as basketID,
count(distinct products.productLabels.product_ProductId) as CountUniqueBasketProducts,
count(distinct products.productLabels.product_VariationId) as CountUniqueBasketVariations
FROM   (SELECT 
        Explode(_ottoaeppoc.productContext.products)   AS products,
        _ottoaeppoc.baseLabels.order_BasketId as basketid
from DataSet_OTTO_IterationFinal2)
group by basketID
limit 5000;
```

```python
data.dtypes
```

```python
data['CountUnqiueBaskets'] = np.where(data['CountUniqueBasketVariations']>data['CountUniqueBasketProducts'], 1, 0)
data
```

```python
%%read_sql data1 -c QS_CONNECTION
select
concat(basketid,products.productLabels.product_ProductId,products.productLabels.product_VariationId) as concatenatedid,
basketid as basketID,
products.productLabels.product_ProductId as productid,
products.productLabels.product_VariationId as variationid,
count(distinct products.productLabels.product_ProductId) as UniqueBasketProducts,
count(distinct products.productLabels.product_VariationId) as UniqueBasketVariations
FROM   (SELECT 
        Explode(_ottoaeppoc.productContext.products)   AS products,
        _ottoaeppoc.baseLabels.order_BasketId as basketid
from DataSet_OTTO_IterationFinal2)
group by concatenatedid,basketID,productid,variationid
order by basketID
limit 5000;
```

```python
data_merged = pd.merge(data,data1,on='basketID',how='left')
data_merged
```
